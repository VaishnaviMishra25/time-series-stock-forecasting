```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.stats.stattools import durbin_watson
from prophet import Prophet
from sklearn.metrics import mean_squared_error, mean_absolute_error
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('HistoricalQuotes.csv')
df.columns = df.columns.str.strip()
df['Date'] = pd.to_datetime(df['Date'])
df.sort_values('Date', inplace=True)
df.set_index('Date', inplace=True)
df = df[['Close/Last', 'Open', 'High', 'Low', 'Volume']]
df.rename(columns={'Close/Last': 'Close'}, inplace=True)
df['Close'] = df['Close'].replace('[\$,]', '', regex=True).astype(float)
```

```python
df
```

```python
df.shape
```

```python
df['Rolling_Mean'] = df['Close'].rolling(window=20).mean()
df['Rolling_Std'] = df['Close'].rolling(window=20).std()
df['Log_Returns'] = np.log(df['Close'] / df['Close'].shift(1))
df['Pct_Returns'] = df['Close'].pct_change()
```

```python
df.sample(10)
```

```python
# Visualizations
plt.figure(figsize=(14, 6))
plt.plot(df['Close'], label='Close Price')
plt.plot(df['Rolling_Mean'], label='Rolling Mean (20 days)')
plt.title('Stock Prices and Rolling Mean')
plt.legend()
plt.show()
```

```python
# Seasonality - monthly average
df['Month'] = df.index.month
monthly_avg = df.groupby('Month')['Close'].mean()
monthly_avg.plot(kind='bar', title='Average Monthly Close Price')
plt.show()
```

```python
decomp = seasonal_decompose(df['Close'], model='multiplicative', period=30)
decomp.plot()
plt.show()
```

```python
df.fillna(method='bfill', inplace=True)

result = adfuller(df['Close'])
print(f'ADF Statistic: {result[0]}')
print(f'p-value: {result[1]}')
```

```python
# If non-stationary, difference the series
df['Close_diff'] = df['Close'].diff()
df.dropna(inplace=True)

result = adfuller(df['Close_diff'])
print(f'ADF after differencing: {result[0]} | p-value: {result[1]}')
```

```python
train_size = int(len(df) * 0.8)
train, test = df.iloc[:train_size], df.iloc[train_size:]
```

```python
model_arima = ARIMA(train['Close'], order=(5,1,2))
model_fit_arima = model_arima.fit()

forecast_arima = model_fit_arima.forecast(steps=len(test))
test['ARIMA_Forecast'] = forecast_arima.values

plt.figure(figsize=(12,6))
plt.plot(train['Close'], label='Train')
plt.plot(test['Close'], label='Test')
plt.plot(test['ARIMA_Forecast'], label='ARIMA Forecast')
plt.legend()
plt.title('ARIMA Forecast')
plt.show()
```

```python
residuals = model_fit_arima.resid

print("Summary statistics of ARIMA residuals:")
print(residuals.describe())
print(f"\nMean of residuals: {residuals.mean()}")
print(f"Durbin-Watson statistic: {durbin_watson(residuals):.4f}")

plt.figure(figsize=(12,6))
plt.subplot(211)
plot_acf(residuals.dropna(), ax=plt.gca(), lags=30)
plt.title('ACF of ARIMA Residuals')

plt.subplot(212)
plot_pacf(residuals.dropna(), ax=plt.gca(), lags=30)
plt.title('PACF of ARIMA Residuals')
plt.tight_layout()
plt.show()
```

```python
df_prophet = df.reset_index()[['Date', 'Close']].rename(columns={'Date': 'ds', 'Close': 'y'})
train_prophet = df_prophet.iloc[:train_size]
test_prophet = df_prophet.iloc[train_size:]

model_prophet = Prophet()
model_prophet.fit(train_prophet)

future = model_prophet.make_future_dataframe(periods=len(test_prophet))
forecast = model_prophet.predict(future)

model_prophet.plot(forecast)
plt.title("Prophet Forecast")
plt.show()
```

```python
# Align forecast with test
forecast_aligned = forecast[['ds', 'yhat']].set_index('ds').join(test[['Close']], how='inner')
test['Prophet_Forecast'] = forecast_aligned['yhat']
```

```python
def evaluate(y_true, y_pred):
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    mae = mean_absolute_error(y_true, y_pred)
    mape = np.mean(np.abs((y_true - y_pred) / y_true)) * 100
    return rmse, mae, mape

arima_eval_df = test[['Close', 'ARIMA_Forecast']].dropna()
prophet_eval_df = test[['Close', 'Prophet_Forecast']].dropna()

arima_rmse, arima_mae, arima_mape = evaluate(arima_eval_df['Close'], arima_eval_df['ARIMA_Forecast'])
prophet_rmse, prophet_mae, prophet_mape = evaluate(prophet_eval_df['Close'], prophet_eval_df['Prophet_Forecast'])

print("ARIMA → RMSE:", arima_rmse, "MAE:", arima_mae, "MAPE:", arima_mape)
print("Prophet → RMSE:", prophet_rmse, "MAE:", prophet_mae, "MAPE:", prophet_mape)
```

```python
# Forecast visualization
plt.figure(figsize=(14,6))
plt.plot(df['Close'], label='Actual')
plt.plot(test.index, test['ARIMA_Forecast'], label='ARIMA Forecast')
plt.plot(test.index, test['Prophet_Forecast'], label='Prophet Forecast')
plt.legend()
plt.title('Forecast Comparison')
plt.show()
```
