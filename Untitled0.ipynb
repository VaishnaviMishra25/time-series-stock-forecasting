```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.stats.stattools import durbin_watson
from prophet import Prophet
from sklearn.metrics import mean_squared_error, mean_absolute_error
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('HistoricalQuotes.csv')
df.columns = df.columns.str.strip()
df['Date'] = pd.to_datetime(df['Date'])
df.sort_values('Date', inplace=True)
df.set_index('Date', inplace=True)
df = df[['Close/Last', 'Open', 'High', 'Low', 'Volume']]
df.rename(columns={'Close/Last': 'Close'}, inplace=True)
df['Close'] = df['Close'].replace('[\$,]', '', regex=True).astype(float)
```

```python
df['Rolling_Mean'] = df['Close'].rolling(window=20).mean()
df['Rolling_Std'] = df['Close'].rolling(window=20).std()
df['Log_Returns'] = np.log(df['Close'] / df['Close'].shift(1))
df['Pct_Returns'] = df['Close'].pct_change()
```

```python
df.sample(10)
```

```python
# Visualizations
plt.figure(figsize=(14, 6))
plt.plot(df['Close'], label='Close Price')
plt.plot(df['Rolling_Mean'], label='Rolling Mean (20 days)')
plt.title('Stock Prices and Rolling Mean')
plt.legend()
plt.show()
```

```python
# Seasonality - monthly average
df['Month'] = df.index.month
monthly_avg = df.groupby('Month')['Close'].mean()
monthly_avg.plot(kind='bar', title='Average Monthly Close Price')
plt.show()
```

```python
decomp = seasonal_decompose(df['Close'], model='multiplicative', period=30)
decomp.plot()
plt.show()
```

```python
df.fillna(method='bfill', inplace=True)

result = adfuller(df['Close'])
print(f'ADF Statistic: {result[0]}')
print(f'p-value: {result[1]}')
```

```python
# If non-stationary, difference the series
df['Close_diff'] = df['Close'].diff()
df.dropna(inplace=True)

result = adfuller(df['Close_diff'])
print(f'ADF after differencing: {result[0]} | p-value: {result[1]}')
```

```python
train_size = int(len(df) * 0.8)
train, test = df.iloc[:train_size], df.iloc[train_size:]
```

```python
model_arima = ARIMA(train['Close'], order=(5,1,2))
model_fit_arima = model_arima.fit()

forecast_arima = model_fit_arima.forecast(steps=len(test))
test['ARIMA_Forecast'] = forecast_arima.values

plt.figure(figsize=(12,6))
plt.plot(train['Close'], label='Train')
plt.plot(test['Close'], label='Test')
plt.plot(test['ARIMA_Forecast'], label='ARIMA Forecast')
plt.legend()
plt.title('ARIMA Forecast')
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
import numpy as np

# Use only the 'Close' price for LSTM
lstm_data = df[['Close']].values

# Normalize the data
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(lstm_data)

# Create sequences (e.g., 60 previous days to predict next day)
def create_sequences(data, seq_len=60):
    X, y = [], []
    for i in range(seq_len, len(data)):
        X.append(data[i-seq_len:i, 0])
        y.append(data[i, 0])
    return np.array(X), np.array(y)

seq_len = 60
X, y = create_sequences(scaled_data, seq_len)

# Reshape X to 3D format for LSTM [samples, time steps, features]
X = X.reshape((X.shape[0], X.shape[1], 1))

# Split into training and testing sets
split = int(0.8 * len(X))
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]
```

```python
model = Sequential()
model.add(LSTM(50, return_sequences=False, input_shape=(X_train.shape[1], 1)))
model.add(Dense(1))

model.compile(optimizer='adam', loss='mean_squared_error')
history = model.fit(X_train, y_train, epochs=20, batch_size=32, validation_data=(X_test, y_test))
```

```python
# Predict
predicted = model.predict(X_test)

# Inverse transform to original scale
predicted_prices = scaler.inverse_transform(predicted.reshape(-1, 1))
actual_prices = scaler.inverse_transform(y_test.reshape(-1, 1))
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error

rmse = np.sqrt(mean_squared_error(actual_prices, predicted_prices))
mae = mean_absolute_error(actual_prices, predicted_prices)
mape = np.mean(np.abs((actual_prices - predicted_prices) / actual_prices)) * 100

print("LSTM â†’ RMSE:", rmse, "MAE:", mae, "MAPE:", mape)
```

```python
plt.figure(figsize=(14,6))
plt.plot(actual_prices, label='Actual')
plt.plot(predicted_prices, label='LSTM Forecast')
plt.legend()
plt.title('LSTM Forecast vs Actual')
plt.show()
```
